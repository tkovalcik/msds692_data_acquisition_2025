```python
import requests
```

GET Request

```python
url = 'https://www.usfca.edu/arts-sciences/programs/graduate/data-science'
headers = ''
r = requests.get(url, headers=headers) # r is a resopnse
print(r.headers)
print(r.content)
```

```python
url = "https://www.google.com/search"
payload = {"q":"USF Data Science"}
headers = ''

r = requests.get(url, params=payload, headers=headers)
print(r.url)
print(r.content)
```

POST Request

```python
url = 'https://www.w3schools.com/action_page.php'
data = {'key': 'value'}
header = ''

r = requests.post(url, data=data, headers=header)
print(r.headers)
print(r.content)
```
